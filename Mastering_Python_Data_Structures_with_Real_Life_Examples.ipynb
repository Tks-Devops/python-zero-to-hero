# Interview Question: List vs Tuple

Here's a comparison between lists and tuples:

| Feature      | List                                    | Tuple                                  |
| :----------- | :-------------------------------------- | :------------------------------------- |
| **Syntax**   | Initialized with square brackets `[]`   | Initialized with parentheses `()`      |
| **Mutability** | Mutable (changeable)                    | Immutable (unchangeable)               |
| **Speed**    | Generally slower than tuples for iteration and access | Generally faster than lists for iteration and access |


## Why is a tuple faster than a list?

In Python, lists are dynamic. When a list grows beyond its allocated memory, Python often creates a new, larger list, copies all the existing elements from the old list to the new one, and then adds the new element. This reallocation and copying process can be time-consuming, especially for large lists.

Tuples, on the other hand, are immutable. Once created, their size and content cannot be changed. This immutability allows Python to allocate memory more efficiently for tuples, often as a single contiguous block. Since there's no need to anticipate future resizing or modifications, operations on tuples can be faster.

## When to choose which data structure?

*   **Choose a `list`** if you need a collection of items that you expect to modify frequently (add, remove, or change elements).
*   **Choose a `tuple`** if you need a collection of items that should remain constant throughout the program's execution, or if you prioritize performance for iteration and access where data immutability is acceptable. Tuples are also often used for heterogeneous data (e.g., a record with different types of information).

The error `TypeError: 'set' object is not subscriptable` means that you cannot access elements in a set using an index, like `set1[0]`. Sets are inherently unordered, so there's no concept of a first, second, or '0th' element. If you need to access elements by position, you might want to consider using a list or tuple instead.

```python
list1 = [2,5,7,2,9,1,2]
## convert the list to set
## to remove the duplicate element
set2 = set(list1)
print(set2)
```

## Common Underlying Data Structure for Sets and Dictionaries
Both sets and dictionaries in Python are implemented using **hash tables** (a form of hashing data structure).

*   **Hashing** allows for very efficient lookups and searches, typically achieving an average time complexity of O(1) (constant time).
*   In a **set**, the elements themselves are hashed to determine their storage location, ensuring uniqueness and fast membership testing.
*   In a **dictionary**, the *keys* are hashed to quickly find their corresponding *values*.
