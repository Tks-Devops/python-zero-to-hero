4) Shallow Copy
list5 = list4.copy()

Now outer list becomes new object.

print(id(list4))
print(id(list5))

IDs will be different.

Memory:

list4 ----> [[1,2],[3,4]]
list5 ----> [[1,2],[3,4]]

But here’s the important part:

Inner lists are still shared

print(id(list4[0]))
print(id(list5[0]))

These IDs will be same.

Modify inner list
list5[0][0] = 100

Now:

print(list4)
print(list5)

Output:

[[100,2],[3,4]]
[[100,2],[3,4]]

Why?

Because inner lists are still shared.

5) Deep Copy

Creates completely independent copy.

import copy

list6 = copy.deepcopy(list4)

Memory:

list4 ----> [[1,2],[3,4]]
list6 ----> [[1,2],[3,4]]

Outer list → new
Inner lists → also new

Modify:

list6[0][0] = 999

Now:

print(list4)
print(list6)

Output:

[[1,2],[3,4]]
[[999,2],[3,4]]

Original stays safe.

Interview Answer
Difference between assignment, shallow copy, and deep copy?
Type	Creates new object?	Nested objects copied?
Assignment (=)	No	No
Shallow copy (copy())	Yes (outer only)	No
Deep copy (deepcopy())	Yes	Yes
Easy rule to remember:
= → same house
.copy() → new house, same furniture
deepcopy() → new house, new furniture

````python
### Graphical Explanation for Laymen

Let's visualize how Python handles lists when you use assignment, shallow copy, and deep copy, using simple diagrams.

---

#### 1. Assignment (`=`): "Same House"

When you assign one list to another (`list2 = list1`), both variables point to the **exact same list object** in memory. Think of it as two different labels pointing to the same physical house.

```
list1
   ↓
   +----------------+
   | [1, 2, 3]      |
   +----------------+
   ↑
list2
```

**What happens if you change something?**
If `list1` adds a new room (element), `list2` will see that new room too, because they are both looking at the same house.

---

#### 2. Shallow Copy (`.copy()`): "New House, Same Furniture"

When you make a shallow copy (`list5 = list4.copy()`), you get a **new outer list object**. This is like building a new house.

However, if your original list contains *other lists* (nested lists), the inner lists (furniture) are **still shared** between the old and new outer lists.

Let's say `list4 = [[1,2], [3,4]]`

```
list4 --------> +----------------+         +----------+
                | [Inner List 1] |-------> | [1, 2]   |
                | [Inner List 2] |-------> | [3, 4]   |
                +----------------+         +----------+

list5 --------> +----------------+         +----------+  <-- Still pointing to the SAME inner lists!
                | [Inner List 1] |-------> | [1, 2]   |
                | [Inner List 2] |-------> | [3, 4]   |
                +----------------+         +----------+
```

**What happens if you change something?**
*   If you add a new inner list to `list4` (add a new piece of furniture to the house), `list5` won't see it because it's a new house.
*   BUT, if you change an element *inside* one of the shared inner lists (e.g., `list5[0][0] = 100`), both `list4` and `list5` will see the change because they share that specific piece of 'furniture' (inner list).

---

#### 3. Deep Copy (`copy.deepcopy()`): "New House, New Furniture"

When you make a deep copy (`list6 = copy.deepcopy(list4)`), you get a **completely independent copy** of the original list. This means not only is the outer list new, but all nested lists (furniture) are also new, separate objects.

Let's say `list4 = [[1,2], [3,4]]`

```
list4 --------> +----------------+         +----------+
                | [Inner List A] |-------> | [1, 2]   |
                | [Inner List B] |-------> | [3, 4]   |
                +----------------+         +----------+


list6 --------> +----------------+         +----------+
                | [Inner List C] |-------> | [1, 2]   |  <-- NEW, independent inner lists
                | [Inner List D] |-------> | [3, 4]   |
                +----------------+         +----------+
```

**What happens if you change something?**
If you modify anything in `list6`, whether it's an outer element or an element deep inside a nested list (`list6[0][0] = 999`), it will *not* affect `list4` at all. They are entirely separate entities, like two completely different houses with completely different furniture.
````
