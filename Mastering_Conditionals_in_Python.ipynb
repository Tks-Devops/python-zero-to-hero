## Typecasting: Changing Data Types

**Typecasting** is the process of converting a value from one data type to another. For example, converting a string that looks like a number into an actual number (integer or float).

**Why is it necessary?**
Python is particular about data types. You can't perform mathematical operations directly on strings, even if they contain digits. If you tried to add '5' (string) and 2 (integer), Python would raise an error. Typecasting allows you to tell Python: "Treat this string '5' as the number 5."

Here, we convert the `percentage` (which was a string) to a `float` so we can compare it with `iit_shortlist_percentage` (which is also a float). Floats are used for numbers with decimal points.

```python
iit_shortlist_percentage = 90.0
candidate_percentage = float(input("enter your percentage: "))

if candidate_percentage >= iit_shortlist_percentage:
  print("congratulations! your selected to an iit college")
else:
  print("sorry, you didn't met the selection criteria")
```

```python
# Demonstrating the difference between if-elif-else and multiple independent if statements

print("--- Approach 1: if-elif-else (Only one block executes) ---")
test_value = 70

if test_value > 90:
    print(f"Condition 1: {test_value} > 90 (Grade A)")
elif test_value > 80:
    print(f"Condition 2: {test_value} > 80 (Grade B)")
elif test_value > 60: # This condition will be met for test_value = 70
    print(f"Condition 3: {test_value} > 60 (Grade C)")
else:
    print(f"Condition 4: {test_value} <= 60 (Grade D)")

print("\n--- Approach 2: Multiple independent if statements (All conditions are checked) ---")
# Using the same test_value = 70
if test_value > 90:
    print(f"Condition 1: {test_value} > 90 (Grade A)")
if test_value > 80:
    print(f"Condition 2: {test_value} > 80 (Grade B)")
if test_value > 60: # This condition will be met
    print(f"Condition 3: {test_value} > 60 (Grade C)")
if test_value <= 60: # This condition will be checked but not met
    print(f"Condition 4: {test_value} <= 60 (Grade D)")

print("\nAs you can see, in Approach 1 (if-elif-else), once a condition is met, the subsequent 'elif' and 'else' blocks are skipped.")
print("In Approach 2 (multiple independent if statements), every 'if' condition is evaluated, even if a previous one was true.")
print("This is why 'if-elif-else' is generally more efficient and appropriate when dealing with a set of mutually exclusive conditions.")
```

## Conditional Logic: Making Decisions

**Conditional logic** allows your program to make decisions and execute different blocks of code based on whether certain conditions are true or false. The `if`, `elif` (else if), and `else` statements are used for this.

### `if-elif-else` vs. Multiple Independent `if` Statements

It's crucial to understand the difference between using an `if-elif-else` chain and using multiple separate `if` statements:

*   **`if-elif-else` (Chained Conditions):**
    *   Only **one** block of code will ever execute within the entire chain.
    *   Once a condition in the `if` or `elif` is met, its corresponding code block runs, and Python **skips all remaining `elif` and `else` conditions** in that chain.
    *   This is ideal for situations where conditions are **mutually exclusive** (e.g., a student can't get both an 'A' and a 'B' grade at the same time). It's also more efficient as it avoids unnecessary checks.

*   **Multiple `if` Statements (Independent Conditions):**
    *   **Every single `if` statement** is evaluated, regardless of whether a previous `if` statement's condition was true.
    *   Multiple blocks of code can execute if their respective conditions are met.
    *   This is used when conditions are **not mutually exclusive**, and you want to check each one independently (e.g., checking if a number is even, and also checking if it's positive – both could be true).

### What is a Palindromic String?

A **palindromic string** (or palindrome) is a sequence of characters that reads the same forwards and backward. This means it remains the same when its characters are reversed.

**Examples:**
*   "mom" (reads the same forwards and backward)
*   "madam"
*   "racecar"
*   "level"

**Non-Examples:**
*   "hello" (reads "olleh" backward)
*   "python"

```python
#Check wheather the character entered by the user is a vowel or a consonant
char = input("enter the character ").lower()
if char == "a" or char == "e" or char == "i" or char == "o" or char == "u":
  print("this char is a vowel")
else:
  print("this is a consonant")
```
